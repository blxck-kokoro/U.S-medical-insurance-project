# U.S. Medical Insurance Costs

In this project we will take a closer look at the dataset of patients, who's information about has been gathered. The main goal is to calculate estimated insurance cost for each patient and focus on analysis of those costs. This data will determine the fundament of our new database - medical records, which has to be clear, legible and flexible in case if some updates will appear. At the end we will analize and sort our own datbase.

At the beginning the csv module should be imported since we'll be working on dataset gathered in file with csv extension.

```python
import csv
```

Now, when we have full acces to csv module, we can read our patients' data. Certain informations about patients will be appended to diffrent lists to make our code more clear.

```python
age_list = []
sex_list = []
bmi_list = []
children_list = []
smoker_list = []
region_list = []
charges_list = []

with open("insurance.csv", "r") as insurance_csv:

    insurance_csv_reader = csv.DictReader(insurance_csv)
    for insurance in insurance_csv_reader:

        age_list.append(insurance["age"])
        sex_list.append(insurance["sex"])
        bmi_list.append(insurance["bmi"])
        children_list.append(insurance["children"])
        smoker_list.append(insurance["smoker"])
        region_list.append(insurance["region"])
        charges_list.append(insurance["charges"])
        
```

Information of all patients from our database has been gathered and appended to certain lists. This allows us to calculate estimated insurance cost for each individual. For this purpose we'll have to process iteration through every list to make sure any element won't be skipped and then use the scheme for calculation.

```python
estimated_insurance_cost_list = []

for age, sex, bmi, children, smoker in zip(age_list, sex_list, bmi_list, children_list, smoker_list):
    if sex == "male":
        sex = 1
    else:
        sex = 0

    if smoker == "yes":
        smoker = 1
    else:
        smoker = 0

    estimated_cost = 250 * int(age )- 128 * int(sex) + 370 * float(bmi) + 425 * int(children) + 24000 * int(smoker) - 12500
    estimated_cost_rounded = round(estimated_cost, 2)
    estimated_insurance_cost_list.append(estimated_cost_rounded)

print(estimated_insurance_cost_list)
```

At this level of research, we are able to calculate averge cost of our medical records and give our patients valuable informations about their insurance costs. In next step we'll focus on displaying message for patient, which will be telling him/her if his/her insurance cost is above or below averge and what should he/she does in case to decrease this cost if wanted.

```python
total_cost = sum(estimated_insurance_cost_list)
costs_ammount = len(estimated_insurance_cost_list)
averge_estimated_insurance_cost = round(total_cost / costs_ammount, 2)

print(f"Averge estimated insurance cost of our medical records so far is: {averge_estimated_insurance_cost}$")
```

```python
for estimated_insurance_cost in estimated_insurance_cost_list:
    if estimated_insurance_cost < averge_estimated_insurance_cost:
        print(f"\nYour estimated insurance cost is {estimated_insurance_cost}$.")

        difference = round(averge_estimated_insurance_cost - estimated_insurance_cost, 2)
        print(f"It is {difference}$ below averge of all gathered in our medical records so far.")

    elif estimated_insurance_cost == averge_estimated_insurance_cost:
        print(f"\nYour esitmated insurance cost is {estimated_insurance_cost}$. It equals averge of all gathered in our medical records so far.")

    elif estimated_insurance_cost > averge_estimated_insurance_cost:
        print(f"\nYour estimated insurance cost is {estimated_insurance_cost}$.")
        
        difference = round(estimated_insurance_cost - averge_estimated_insurance_cost, 2)
        print(f"It is {difference}$ above averge of all gathered in our medical records so far.")
```

```python
def analyze_bmi(bmi_value):
  if bmi_value > 30:
    return('Your BMI is in the obese range. To lower your cost, you should significantly lower your BMI.')
  elif bmi_value >= 25:
    return ('Your BMI is in the overweight range. To lower your cost, you should lower your BMI.')
  elif bmi_value >= 18.5 and bmi_value < 25:
    return('Your BMI is in a healthy range.')
  elif bmi_value < 18.5:
    return('Your BMI is in the underweight range. Increasing your BMI will not help lower your cost, but it will help improve your health.')

def bmi_advice(bmi_value):
  if bmi_value > 25:
      bmi_reduction = bmi_value - 25
      return(f'You should reduce your BMI level by {bmi_reduction} in order to keep your BMI healthy.')
  if bmi_value < 18.5:
      bmi_increase = 18.5 - bmi_value
      return(f'You should increase your BMI level by {bmi_increase} in order to keep your BMI healthy.')
  if bmi_value < 25 and bmi_value > 18.5:
      return

import decimal

for bmi in bmi_list:
    bmi = decimal.Decimal(bmi)

    print(f"Analyze: {analyze_bmi(round(bmi, 2))}")
    print(f"Advice: {bmi_advice(round(bmi, 2))}")
    print("\n")
```

```python
def analyze_smoker(smoker_status):
  if smoker_status == 1:
    return('To lower your cost, you should consider quitting smoking.')
  else:
    return('Smoking is not an issue for you.')

for smoker in smoker_list:
    if smoker == "yes":
        smoker = 1
    else:
        smoker = 0
    
    print(analyze_smoker(smoker))
    print("\n")
```

We've succefully extended our database with valuable data of our patients in medical records. Now we can take next step forward and create possibility to build patient's individual profile, basing on his own data. We'll be using python classes to make them clear and simple as much as possible. Also we will write some methods, which will provide possibility for our clients to update their data if needed.

```python
class Patient:

    def __init__(self, first_name, last_name, age, sex, bmi, num_of_children, smoker, region, charges, estimated_insurance_cost):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

        if sex == "male":
            sex = 1
        else:
            sex = 0

        self.sex = sex
        self.bmi = bmi
        self.num_of_children = num_of_children

        if smoker == "yes":
            smoker = 1
        else:
            smoker = 0

        self.smoker = smoker
        self.region = region 
        self.charges = charges 

        self.estimated_insurance_cost = estimated_insurance_cost

    def estimated_insurance_cost_calculation(self):
        estimated_cost = 250 * int(self.age)- 128 * int(self.sex) + 370 * float(self.bmi) + 425 * int(self.num_of_children) + 24000 * int(self.smoker) - 12500
        print(f"{self.first_name} {self.last_name}'s estimated insurance cost is now {estimated_cost} dollars.")

    
    def update_age(self, new_age):
        self.age = new_age
        print(f"{self.first_name} {self.last_name} is now {new_age} years old.")
        self.estimated_insurance_cost_calculation()

    def update_num_children(self, new_num_children):
        self.num_of_children = new_num_children
        if new_num_children == 1:
            print(f"{self.first_name} {self.last_name} has {new_num_children} child.")
        else:
            print(f"{self.first_name} {self.last_name} has {new_num_children} children.")
        self.estimated_insurance_cost_calculation()

    def update_bmi(self, new_bmi):
        self.bmi = new_bmi
        print(f"{self.first_name} {self.last_name}'s new bmi is {new_bmi}.")
        self.estimated_insurance_cost_calculation()

    def update_smoking_status(self, new_smoking_status):
        self.smoker = new_smoking_status
        if new_smoking_status == 0:
            print(f"{self.first_name} {self.last_name} isn't smoker anymore.")
        elif new_smoking_status == 1:
            print(f"{self.first_name} {self.last_name} is now smoker.")
        self.estimated_insurance_cost_calculation()
  
    def patient_profile(self):
        patient_informations = {
        "First Name":self.first_name,
        "Last Name":self.last_name,
        "Age":self.age,
        "Sex":self.sex,
        "BMI":self.bmi,
        "Number Of Children":self.num_of_children,
        "Smoker":self.smoker,
        "Region":self.region,
        "Charges":self.charges,
        "Estimated Insurance Cost":self.estimated_insurance_cost
        }

        for patient_information_key, patient_information_value in patient_informations.items():
            print (f"{patient_information_key} : {patient_information_value}")



patient_0 = Patient(
    first_name = "Patient",
    last_name = "Zero",
    age = age_list[0], 
    sex = sex_list[0],
    bmi = bmi_list[0], 
    num_of_children = children_list[0],
    smoker = smoker_list[0],
    region = region_list[0],
    charges = charges_list[0],
    estimated_insurance_cost = estimated_insurance_cost_list[0]
    )

patient_0.patient_profile()
print("\n")

patient_0.update_age(20)
print("\n")
patient_0.update_bmi(28.0)
print("\n")
patient_0.update_num_children(1)
print("\n")
patient_0.update_smoking_status(0)

```

# Insurance cost analysis

```python
highest_estimated_insurance_cost = max(estimated_insurance_cost_list)
print(f"The highest estimated insurance cost in our database: {highest_estimated_insurance_cost}$")

lowest_estimated_insurance_cost = min(estimated_insurance_cost_list)
print(f"The lowest estimated insurance cost in our database: {lowest_estimated_insurance_cost}$")

total_cost = sum(estimated_insurance_cost_list)
costs_ammount = len(estimated_insurance_cost_list)
averge_estimated_insurance_cost = round(total_cost / costs_ammount, 2)
print(f"Averge estimated insurance cost in our database: {averge_estimated_insurance_cost}$")
```

# Age analysis

```python
eldest_patient = max(age_list)
print(f"The eldest patient registered in database is {eldest_patient} years old.")

youngest_patient = min(age_list)
print(f"The youngest patient registered in database is {youngest_patient} years old.")

total_age = 0
for age in age_list:
    total_age += int(age)

age_ammount = len(age_list)
averge_age = round(total_age / age_ammount, 0)
print(f"Averge age in database: {averge_age}")
```

# Sex analysis

```python
males_ammount = 0
females_ammount = 0

for sex in sex_list:
    if sex == "male":
        males_ammount += 1
    elif sex == "female":
        females_ammount += 1

print(f"Ammount of males in database: {males_ammount}")
print(f"Ammount of females in database {females_ammount}")
    
if males_ammount > females_ammount:
    difference = males_ammount - females_ammount
    print(f"There is {difference} more males than females in our database.")
elif females_ammount > males_ammount:
    difference = females_ammount - males_ammount
    print(f"There is {difference} more females than males in our database.")
elif males_ammount == females_ammount:
    print("Ammount of males and females in our database is on the same level.")
```

# Bmi analysis

```python
highest_bmi = max(bmi_list)
lowest_bmi = min(bmi_list)

total_bmi = 0
for bmi in bmi_list:
    total_bmi += float(bmi)

bmi_ammount = len(bmi_list)

averge_bmi = round(total_bmi / bmi_ammount, 2)

print(f"The highest bmi in database: {highest_bmi}")
print(f"The lowest bmi in database: {lowest_bmi}")
print(f"Averge bmi in database: {averge_bmi}")
```

# Children analysis

```python
highest_children_ammount_per_parent = 0
lowest_children_ammount_per_parent = 0

for children in children_list:
    if int(children) > highest_children_ammount_per_parent:
        highest_children_ammount_per_parent = int(children)

    if int(children) < lowest_children_ammount_per_parent:
        lowest_children_ammount_per_parent = int(children)


total_children_ammount = 0

for children in children_list:
    total_children_ammount += int(children)



parents = len(children_list)
averge_children_ammount = round(total_children_ammount / parents, 0)

print(f"The highest ammount of children per parent in database: {highest_children_ammount_per_parent}")
print(f"The lowest ammount of children per parent in database: {lowest_children_ammount_per_parent}")
print(f"Averge ammount of children per parent in database: {averge_children_ammount}")
```

# Smokers analysis

```python
smokers = 0
none_smokers = 0

for smoker in smoker_list:
    if smoker == "yes":
        smokers += 1
    elif smoker == "no":
        none_smokers += 1

print(f"Ammount of smokers in database: {smokers}")
print(f"Ammount of none smokers in database {none_smokers}")

if smokers > none_smokers:
    difference = smokers - none_smokers
    print(f"There is {difference} more smokers than none smokers in our database.")
elif none_smokers > smokers:
    difference = none_smokers - smokers
    print(f"There is {difference} more none smokers than smokers in our database.")
elif smokers == none_smokers:
    print(f"Ammount of smokers and none smokers in our database is at the same level.")


```

# Region analysis

```python
region_names = set(region_list)

print(f"List of regions where our patients are from:")
for region in region_names:
    print(f"-{region.title()}")
print("\n")

print(f"Number of patients from NorthEast region: {region_list.count('northeast')}")
print(f"Number of patients from NorthWest region: {region_list.count('northwest')}")
print(f"Number of patients from SouthEast region: {region_list.count('southeast')}")
print(f"Number of patients from SouthWest region: {region_list.count('southwest')}")





```

# Charges analysis

```python
float_charges_list = []
for charge in charges_list:
    float_charges_list.append(float(charge))

highest_charges = round(max(float_charges_list), 2)
lowest_charges = round(min(float_charges_list), 2)

total_charges = 0
for charges in charges_list:
    total_charges += float(charges)
charges_ammount = len(charges_list)
averge_charges = round(total_charges / charges_ammount, 2)

print(f"Highest charges record in database: {highest_charges}$")
print(f"Lowest charges record in database: {lowest_charges}$")
print(f"Averge charges record in database: {averge_charges}$")
```
